# 2. Bayesian optimization using GPyOpt

The Python code that was used to generate the results for the document is summarized here; instead of GPy, GPyOpt was used, as indicated in the first notebook.

First, introduce the packages we need (make sure to have them installed):

```python
import os
import GPyOpt
import time
import numpy as np
from numpy.random import seed  
seed(123456) # fixed seed
start_time = time.time() # For timekeeping
```

Then, some initial parameters can be specified.

```python
max_iter = 20  # Max number of iterations

space = [{'name': 'var_1', 'type': 'continuous', 'domain': (30.0, 34.18)},
         {'name': 'var_2', 'type': 'continuous', 'domain': (50.0, 200.0)}]

# In- and output text files
a_Design_var1_txt_file = "a_Design_var1.txt"
a_Design_var2_txt_file = "a_Design_var2.txt"
b_Objective_1_txt_file = "b_Objective_c_gap.txt"

# Command line to run ABAQUS
cmdl = "abaqus cae noGUI=2D-get.py"
```

Next, we define the gap size objective function to minimize.

```python
def gap_size(x):
    open(a_Design_var1_txt_file, "w").write(str(x[:, 0][0]))
    open(a_Design_var2_txt_file, "w").write(str(x[:, 1][0]))
    # os.system(cmdl)
    return float(open(b_Objective_1_txt_file, "r").read().strip())
```

The optimization loop can now be initialized. This loop will also output the plotted optimization history.

```python
current_iter = 0

files = []  # File name string array initialization

bo_step = GPyOpt.methods.BayesianOptimization(f=gap_size, domain=space)  # Initial DOE

X_step = bo_step.X  # Defining input of iterative step
Y_step = bo_step.Y  # Defining output of iterative step

while current_iter <= max_iter:
    bo_step = GPyOpt.methods.BayesianOptimization(f=gap_size, domain=space, X=X_step, Y=Y_step)  # Optimization step
    x_next = bo_step.suggest_next_locations()  # Querying new optimization input point
    y_next = gap_size(x_next)

    X_step = np.vstack((X_step, x_next))  # Combining older steps with new step
    Y_step = np.vstack((Y_step, y_next))

    # Output iteration progress by printing the relevant data
    current_iter += 1
    print("iteration ", current_iter)
    print("time :", time.time()-start_time)
    print("input :",  X_step)
    print("output :", Y_step)

    # Save images into this folder
    if current_iter < 10:
        strxFile = 'opt_0%d' % current_iter
    else:
        strxFile = 'opt_%d' % current_iter
    files.append(strxFile + ".png")
    bo_step.plot_acquisition(filename=strxFile)

bo_step.save_evaluations(evaluations_file = "GPyOpt_obj.txt")
```
